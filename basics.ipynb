### import packages

````python
import pandas as pd
from muselsl import view
from matplotlib import pyplot as plt
import matplotlib
````

````python
matplotlib.rcParams["figure.figsize"] = [15, 8]
````

### start and record stream from command line

```>>> muselsl stream -a 00:55:DA:B5:85:4C```

```>>> muselsl record -d 60 -f ./<file_name>.csv```

### visualize real-time recording

````python
view(version=1, backend="nbAgg", figure='15x10')
````

### read recorded data

````python
data = pd.read_csv("test_09_30.csv")
````

````python
channel_dict = {"AF7": "red", "AF8": "orange", "TP9": "green", "TP10": "blue"}
````

````python
for channel, color in channel_dict.items():

    plt.plot(
        data.loc[5000:6000, "timestamps"].values,
        data.loc[5000:6000, channel].values,
        c=color,
        label=channel,
    )

plt.title("EEG data")
plt.legend()
````
